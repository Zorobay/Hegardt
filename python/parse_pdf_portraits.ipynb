```python
from pathlib import Path
import cv2
from matplotlib import pyplot as plt
import numpy as np

PAGES_DIR = Path('./pages_as_images')
OUTPUT_DIR = Path('./contour_portraits')
OUTPUT_DIR.mkdir(parents=True, exist_ok=True)


def show_rgb_image(image, title=None, conversion=cv2.COLOR_BGR2RGB):
    # Converts from one colour space to the other. this is needed as RGB
    # is not the default colour space for OpenCV
    image = cv2.cvtColor(image, conversion)
    plt.imshow(image)

    # remove the axis / ticks for a clean looking image
    plt.xticks([])
    plt.yticks([])

    # if a title is provided, show it
    if title is not None:
        plt.title(title)

    plt.show()
```

```python
page1 = list(PAGES_DIR.glob('*.png'))[0]

# Read page
img = cv2.imread(str(page1))
show_rgb_image(img, 'Read page')
```

```python
# Convert to grayscale
gray = cv2.cvtColor(img, cv2.COLOR_BGR2GRAY)
show_rgb_image(gray, 'Grayscale')
```

```python
# Threshold the image -> anything above light gray becomes white, the rest becomes black (frames are light gray)
_, thresh = cv2.threshold(gray, 230, 250, cv2.THRESH_BINARY_INV)
show_rgb_image(thresh)
```

```python
# Find contours
contours, _ = cv2.findContours(thresh, cv2.RETR_EXTERNAL, cv2.CHAIN_APPROX_SIMPLE)
contour_areas = [cv2.contourArea(cont) for cont in contours]
max_index = int(np.argmax(contour_areas))
max_contour = contours[max_index].reshape(-1,2)
#show_rgb_image(contours[max_index], 'Contours', cv2.COLOR_BGR2GRAY)
page = img.copy()
for x,y in max_contour:
    cv2.circle(page, (x, y), 1, (255, 0, 0), 3)
show_rgb_image(page)
```
